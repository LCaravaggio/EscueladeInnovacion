# Partición Train Test

```python
from vega_datasets import data
cars=data.cars()
cars
```

```python
from sklearn.model_selection import train_test_split
col_X=['Miles_per_Gallon',	'Cylinders',	'Displacement', 'Horsepower'	,'Weight_in_lbs',	'Acceleration']
col_Y=['Origin']
X_train, X_test, y_train, y_test = train_test_split(cars[col_X], cars[col_Y], test_size=0.3, random_state=42, shuffle=True)
```

# Distancia Euclidiana

```python
import numpy as np

point1 = X_train.iloc[1,].values
point2 = X_train.iloc[2,].values

dist = np.linalg.norm(point1 - point2)

print(dist)
```

```python
import math

print(math.dist(point1, point2))
```

```python
def euclidean_distance(point1, point2):
    diff = point1 - point2
    distance = np.sqrt(np.sum(diff**2))
    return distance

euclidean_distance(point1, point2)
```

```python
# Análisis univariado
# Histograma para cada característica numérica

import matplotlib.pyplot as plt
X_train.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# Análisis bivariado
# Matriz de correlación
import seaborn as sns

correlation_matrix = X_train.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Gráficos de dispersión (scatter plots) para algunas características
sns.pairplot(X_train, diag_kind='kde')
plt.show()
```

```python
# Análisis de distribución
# Usando seaborn para visualizar la distribución de algunas columnas
for column in X_train.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(5, 3))
    sns.histplot(X_train[column], kde=True)
    plt.title(f'Distribución de {column}')
    plt.show()
```

```python
# Boxplots para detectar valores atípicos
for column in X_train.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(5, 3))
    sns.boxplot(x=X_train[column])
    plt.title(f'Boxplot de {column}')
    plt.show()
```
