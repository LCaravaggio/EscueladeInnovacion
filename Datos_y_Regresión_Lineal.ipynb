# Regresión lineal

```python
import statsmodels.api as sm
import numpy as np
```

```python
X = pd.DataFrame({'intercept': np.ones(Auto.shape[0]),
                  'horsepower': Auto['horsepower']})
```

```python
y = Auto['mpg']
model = sm.OLS(y, X)
results = model.fit()
```

```python
results.summary()
```

```python
X = pd.DataFrame({'intercept': np.ones(Auto.shape[0]),
                  'horsepower': Auto['horsepower'],
                  'year': Auto['year'],
                  'displacement': Auto['displacement']})
```

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```
