# Tabla Interactiva

```python
from vega_datasets import data
```

```python
cars=data.cars()
cars
```

# SQL

```python
import sqlite3

database_path = 'cars.db'
con = sqlite3.connect(database_path)

# Convierto el DF a una base de SQL
cars.to_sql('cars_table', con, if_exists='replace', index=False)

# Y listo, puedo ejecutar instrucciones
cur = con.cursor()
cur.execute("SELECT * FROM cars_table WHERE Miles_per_Gallon > 40")
rows = cur.fetchall()

for row in rows:
    print(row)

# Cierro la conexión
con.close()
```
